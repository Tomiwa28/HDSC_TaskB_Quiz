Answer to Question 14 - 20

```python
energy_data_2= pd.read_csv('energydata_complete (1).csv')
energy_data_2['date'] = pd.to_numeric(energy_data_2['date'], errors='coerce')
```

Normalizing the dataset using the MinMaxScaler after removing the following columns: [“date”, “lights”]

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
normalised_energy_data = pd.DataFrame(scaler.fit_transform(energy_data_2), columns=energy_data_2.columns)
features_energy_data = normalised_energy_data.drop(columns=['date', 'lights'])
energy_data_target = normalised_energy_data['Appliances']
```

Splitting the dataset

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 0)
```

Running a multiple linear regression using the training set and evaluating the model on the test set.


```python
from sklearn.linear_model import LinearRegression
reg2 = LinearRegression()
reg2.fit(x_train, y_train)
```

```python
y_pred = reg2.predict(x_test)
y_pred
```

Mean Absolute Error (in two decimal places)

```python
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(y_test, y_pred)
round(mae, 2) 
```

Residual Sum of Squares (in two decimal places)

```python
from sklearn.metrics import r2_score
value = r2_score(y_test, y_pred)
round(value, 2)
```

Root Mean Squared Error (in three decimal places)

```python
from sklearn.metrics import mean_squared_error
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
round(rmse, 3) 
```

Training a ridge regression model with an alpha value of 0.4

```python
from sklearn.linear_model import Ridge
ridge_reg = Ridge(alpha=0.4)
ridge_reg.fit(x_train, y_train)
```

```python
y_pred2 = ridge_reg.predict(x_test)
y_pred2
```

Changes in the RMSE

```python
from sklearn.metrics import mean_squared_error
rmse = np.sqrt(mean_squared_error(y_test, y_pred2))
round(rmse, 3) 
```

Training a lasso regression model with an alpha value of 0.001

```python
from sklearn.linear_model import Lasso
lasso_reg = Lasso(alpha=0.001)
lasso_reg.fit(x_train, y_train)
```

```python
y_pred3 = lasso_reg.predict(x_test)
y_pred3
```

RMSE with the lasso regression

```python
from sklearn.metrics import mean_squared_error
rmse = np.sqrt(mean_squared_error(y_test, y_pred3))
round(rmse, 3) 
```
